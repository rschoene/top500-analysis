```python
from collections import Counter
import matplotlib.pyplot as plt

df = pd.read_excel("original/TOP500_202311.xlsx")
element_counts = Counter(df["Interconnect Family"])

labels=list(element_counts.keys())

fig, ax = plt.subplots()
ax.pie(element_counts.values(), labels=element_counts.keys())
ax.set_title("Network System Share")

values=[]
new_sum=df["Rmax [TFlop/s]"].sum()
for element in labels:
    values.append(0)
for index, row in df.iterrows():
    values[labels.index(row["Interconnect Family"])]+=row["Rmax [TFlop/s]"]*100.0/new_sum
    
fig, ax = plt.subplots()
ax.pie(values, labels=labels)
ax.set_title("Network Performance Share")
```

```python
from collections import Counter
import matplotlib.pyplot as plt

df = pd.read_excel("original/TOP500_202311.xlsx")

df.loc[df['Accelerator/Co-Processor'].str.contains("NVIDIA"), 'Accelerator/Co-Processor'] = "NVIDIA"
df.loc[df['Accelerator/Co-Processor'].str.contains("Nvidia"), 'Accelerator/Co-Processor'] = "NVIDIA"
df.loc[df['Accelerator/Co-Processor'].str.contains("AMD"), 'Accelerator/Co-Processor'] = "AMD"
df.loc[df['Accelerator/Co-Processor'].str.contains("Xeon Phi"), 'Accelerator/Co-Processor'] = "Intel Xeon Phi"
df.loc[df['Accelerator/Co-Processor'].str.contains("Intel Data Center GPU"), 'Accelerator/Co-Processor'] = "Intel Data Center GPU"


element_counts = Counter(df["Accelerator/Co-Processor"])

labels=sorted(list(element_counts.keys()))
predefined_start = ["NVIDIA","AMD", "Intel Data Center GPU", "Intel Xeon Phi"]
predefined_end = ["None"]
for movelabel in reversed(predefined_start):
    labels.remove(movelabel)
    labels.insert(0,movelabel)
    
for movelabel in predefined_end:
    labels.remove(movelabel)
    labels.append(movelabel)

counts_sorted=[]
for label in labels:
    counts_sorted.append(element_counts[label])


fig, ax = plt.subplots()
print(element_counts.values())
ax.pie(counts_sorted)
plt.legend(labels, loc='center left', bbox_to_anchor=(1, 0.5))
ax.set_title("Accelerator System Share")
plt.show()

values=[]
new_sum=df["Rmax [TFlop/s]"].sum()
for element in labels:
    values.append(0)
for index, row in df.iterrows():
    values[labels.index(row["Accelerator/Co-Processor"])]+=row["Rmax [TFlop/s]"]*100.0/new_sum
    
print(values)
fig, ax = plt.subplots()
ax.pie(values)
plt.legend(labels, loc='center left', bbox_to_anchor=(1, 0.5))
ax.set_title("Accelerator Performance Share")
plt.show()
```
